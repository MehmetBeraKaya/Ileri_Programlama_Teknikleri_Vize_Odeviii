```python
import math
import random
import matplotlib.pyplot as plt

# Okul Numaram 22040201012
okul_numarasi_son_2_hane = 12
top_konum = [0, okul_numarasi_son_2_hane]

aci = 30  
aci_radyan = math.radians(aci)

uzaklik_mesafesi = 20000 + 200 * random.randint(-10, 10)
genislik_baslangic = uzaklik_mesafesi
genislik_bitis = uzaklik_mesafesi + 1000 + 100 * random.randint(-2, 2)

hiz_alt_sinir = 330
hiz_ust_sinir = 1800
hiz = (hiz_alt_sinir + hiz_ust_sinir) / 2

atis_sayisi = 0
hedef_vuruldu = False

def topun_dusme_noktasi(hiz, aci_radyan, top_konum):
    yatayhiz=hiz*math.cos(aci_radyan)
    duseyhiz=hiz*math.sin(aci_radyan)
    ucussuresi=(2*duseyhiz)/(9.8)
    menzil=yatayhiz*ucussuresi
    duseyyol=duseyhiz*ucussuresi-(1/2)*(9.8)*ucussuresi**2
    x_koordinati = top_konum[0] + menzil
    y_koordinati = top_konum[1]+duseyyol
    return (x_koordinati, y_koordinati)

while not hedef_vuruldu:
    atis_sayisi += 1
    dusme_noktasi = topun_dusme_noktasi(hiz, aci_radyan, top_konum)

    if genislik_baslangic <= dusme_noktasi[0] <= genislik_bitis:
        hedef_vuruldu = True
        print(f"Hedef {atis_sayisi}. atişta vuruldu. Gerekli hiz: {hiz} m/s")
        plt.plot(dusme_noktasi[0],dusme_noktasi[1])
        plt.title(f"Deneme sayisi = {atis_sayisi}")
        plt.xlabel("Mesafe (m)")
        plt.ylabel("Yükseklik (m)")
        plt.xlim(0, 25000)
        plt.show()

    else:
        if dusme_noktasi[0] < genislik_baslangic:
            hiz_alt_sinir = hiz
            print(f"{atis_sayisi}. atişta top mermisi hedefin önüne düştü.")
        else:
            hiz_ust_sinir = hiz
            print(f"{atis_sayisi}. atişta top mermisi hedefin arkasina düştü.")
        hiz = (hiz_alt_sinir + hiz_ust_sinir) / 2
```
